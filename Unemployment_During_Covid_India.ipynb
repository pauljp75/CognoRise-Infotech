**OBJECTIVE**

**Dataset Overview:**


*   Measuring Frequency: The regularity of measurement collection (Monthly)
*   Estimated Unemployment Rate (%): The proportion of unemployed individuals  
    in each Indian state.
*   Estimated Employed Individuals: The total count of currently engaged individuals.
*   Estimated Labour Participation Rate (%): The percentage of the working-age population actively involved in the job market, encompassing both employed individuals and those actively seeking employment.








This dataset facilitates an understanding of the evolving unemployment patterns among India's states amidst the COVID-19 crisis, providing insights into the nuanced impacts on unemployment rates, employment metrics, and labor participation rates across diverse regions. The analysis aims to illuminate the socio-economic repercussions of the pandemic on India's workforce and labor market.

```python
#rename columns
df.columns=['State', 'Date', 'Frequency', 'Estimated Unemployment Rate',
       'Estimated Employed', 'Estimated Labour Participation Rate',
       'Region', 'longitude', 'latitude']
```

```python
#changing datatype of date to datetime
df['Date']=pd.to_datetime(df['Date'],dayfirst=True)
```

```python
df.dtypes
```

```python
#Adding a month column for better analysis based on months

#isolate month from date
df['monthvalue']=df['Date'].dt.month

df['Month']=df['monthvalue'].apply(lambda x: calendar.month_abbr[x])

df.drop(['monthvalue'],axis=1)
```

# **Monthly Analysis**

```python
#Group by Month and Mean of Numeric Values
data=df.groupby(['Month'])[['Estimated Unemployment Rate','Estimated Employed', 'Estimated Labour Participation Rate']].mean()
data=pd.DataFrame(data).reset_index()
data
```

**Unemployment Rate and Labour Participation Rate by Month**

```python
fig=go.Figure()

fig.add_trace(go.Bar(x=data['Month'],y=data['Estimated Unemployment Rate'],name='Unemployment Rate'))
fig.add_trace(go.Bar(x=data['Month'],y=data['Estimated Labour Participation Rate'],name='Labour Participation Rate'))

fig.update_layout(title='Unemployment Rate and Labour Participation Rate by Month',xaxis={'categoryorder':'array','categoryarray':['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct']})

fig.show()
```

**Estimated Employed per Month**

```python
fig=go.Figure()

fig.add_trace(go.Bar(x=data['Month'],y=data['Estimated Employed'],name='Employed'))


fig.update_layout(title='Estimated Employed by Month',xaxis={'categoryorder':'array','categoryarray':['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct']})

fig.show()
```

# **Unemployment Rate Before and After Lockdown**

```python
#get month value
df['monthvalue']=df['Date'].dt.month


#split before and after lockdown
before_lockdown=df[(df['monthvalue']>=1) &(df['monthvalue'] <4)]
after_lockdown=df[(df['monthvalue'] >=4) & (df['monthvalue'] <=6)]



#Calculating the mean unemployment rate before lockdown by state
after_ld=after_lockdown.groupby('State')['Estimated Unemployment Rate'].mean().reset_index()
#Calculating the mean unemployment rate after lockdown by state
before_ld=before_lockdown.groupby('State')['Estimated Unemployment Rate'].mean().reset_index()



#Combining the mean unemployment rates before and after lockdown by state
df2=pd.merge(before_ld,after_ld,on='State')
df2.columns=['State','Unemployment Rate Before Lockdown','Unemployment Rate After Lockdown']




```

```python
df2
```

**Unemployment Rate Change Before and After Lockdown**

```python
df2['Unemployment Rate Change']=round(df2['Unemployment Rate After Lockdown']-df2['Unemployment Rate Before Lockdown']
                         /df2['Unemployment Rate After Lockdown'],2)
df2.head()
```

```python
fig=px.bar(df2,x='State',y='Unemployment Rate Change',color='Unemployment Rate Change',
            title='Unemployment Rate Change by State After lockdown')#,template='ggplot2')
fig.update_layout(xaxis={'categoryorder':'total ascending'})
fig.show()
```

# **Conclusion**


The dataset played a pivotal role in our comprehensive analysis, shedding light on the profound impact of the Covid-19 lockdown on employment across India. Our investigation delved into the dynamic shifts and trends that unfolded in the wake of the lockdown, revealing a nuanced picture. Notably, the Northern states bore a more substantial brunt compared to their Southern counterparts.

The initial wave of unemployment surged nationwide, with April and May recording the highest unemployment rates. However, as the year progressed, we observed a gradual decline in these rates, culminating in a relative stabilization by October.

Region-specific insights underscored the disproportionate effects, with Haryana and Tripura emerging as the states grappling with the highest unemployment rates. Additionally, our findings spotlighted Puducherry, where the rate of unemployment experienced the most significant surge when comparing pre and post-lockdown data.

This nuanced analysis, grounded in regional and temporal dimensions, not only enhances our understanding of the immediate repercussions but also serves as a valuable resource for policymakers and stakeholders navigating the complex landscape of post-lockdown economic recovery in India.
